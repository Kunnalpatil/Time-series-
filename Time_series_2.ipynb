```python
df.head()
```

# **Resample Rules**



            B         business day frequency
            C         custom business day frequency (experimental)
            D         calendar day frequency
            W         weekly frequency
            M         month end frequency
            SM        semi-month end frequency (15th and end of month)
            BM        business month end frequency
            CBM       custom business month end frequency
            MS        month start frequency
            SMS       semi-month start frequency (1st and 15th)
            BMS       business month start frequency
            CBMS      custom business month start frequency
            Q         quarter end frequency
            BQ        business quarter endfrequency
            QS        quarter start frequency
            BQS       business quarter start frequency
            A         year end frequency
            BA, BY    business year end frequency
            AS, YS    year start frequency
            BAS, BYS  business year start frequency
            BH        business hour frequency
            H         hourly frequency
            T, min    minutely frequency
            S         secondly frequency
            L, ms     milliseconds
            U, us     microseconds
            N         nanoseconds

```python
df.resample(rule='A').min() # To get data of last date of every year
```

```python
df.resample(rule='A').max()['Open'].plot()
```

```python
# Quaterly data
df.resample(rule='QS').max()
```

```python
df.resample(rule='QS').max()['High'].plot()
```

```python
df.resample(rule='BA').max()['High'].plot(kind='bar')
```

```python
df.resample(rule='M').mean()['High'].plot(kind='bar',figsize=(16,8)) #Montly mean
```

# **Simple moving average**

```python
df['Open: 30 days rolling mean']=df['Open'].rolling(30).mean()
```

```python
df[['Open','Open: 30 days rolling mean']].plot(xlim=('2020','2022'),figsize=(16,8))
```

```python
df['Open:10 days rolling']=df['Open'].rolling(window=10,min_periods=2).mean()
```

```python
df.head()
```

```python
df[['Open','Open: 30 days rolling mean','Open:10 days rolling']].plot(xlim=('2020','2022'),figsize=(20,12))
```

# Cumulative moving average

```python

df['Cumulative average']=df['Open'].expanding().mean().plot()
```

```python
df['Cumulative average'].plot(figsize=(16,5))
df['Open: 30 days rolling mean'].plot()
df['Open'].plot()
```

# **Exponential Moving average**
## To prevent lags 

formula : ((column - prev EMA)*multiplier)+prev EMA

multiplier formula : 2/(10+1) 10: rolling ema size

initially start with simple moving average as prev ema 



```python
df['EMA_0.1']=df['Open'].ewm(alpha=0.1,adjust=False).mean() # Alpha is smoothing factor 
```

```python
df['EMA_0.3']=df['Open'].ewm(alpha=0.3,adjust=False).mean()
```

```python
df[['Open','EMA_0.1','EMA_0.3']].plot(xlim=('2020','2022'),figsize=(20,8))
```

# **Exponential Weighted Moving average** 
## (BEST Technique)

## To prevent lags 

```python
df['EWMA_5days']=df['Open'].ewm(span=5).mean()
```

```python
df[['Open','EMA_0.1','EWMA_5days']].plot(xlim=('2021-6-25','2022'),figsize=(20,8))
```
