## Ford and Gm Related (Positive correlation)

# **Daily percentage change**

```python
tesla['returns']= (tesla['Close']/tesla['Close'].shift(1))-1 # Closing price / closing price of previous data -1
```

```python
tesla.head()
```

```python
gm['returns']= (gm['Close']/gm['Close'].shift(1))-1 
ford['returns']= (ford['Close']/ford['Close'].shift(1))-1 
```

# **Volatility**

The spreaded(fat) the histogram, the more volatility



```python
ford['returns'].hist(bins=50) 
```

```python
gm['returns'].hist(bins=50)
```

```python
tesla['returns'].hist(bins=50)
```

```python
tesla['returns'].hist(bins=100,label='Tesla',alpha=0.6,figsize=(14,6)) # alpha for transparence
gm['returns'].hist(bins=100,label='GM',alpha=0.4)
ford['returns'].hist(bins=100,label='Ford',alpha=0.5)
plt.legend()
```

# Tesla more volatile

```python
# Smooth the hist (Probablity distribution function)
```

```python
tesla['returns'].plot(kind='kde',label='Tesla',figsize=(15,6))
gm['returns'].plot(kind='kde',label='Tesla')
ford['returns'].plot(kind='kde',label='Tesla')
plt.legend()
```

Tesla is more volatile , ford and gm are almost similar 

#**BOX PLOTS**

```python
box_df = pd.concat([tesla['returns'],gm['returns'],ford['returns']],axis=1)
box_df.columns=['Tesla returns','GM Returns','Ford returns']
box_df.plot(kind='box',figsize=(12,6))
```

```python
scatter_matrix(box_df,figsize=(8,8),hist_kwds={'bins':50})
```

GM and Ford returns are correlated
